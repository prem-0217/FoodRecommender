```python
numeric = ['protein', 'carbs', 'fat']
df.boxplot(numeric)
```

```python
df.boxplot('calories')
```

```python
fatmean = df['fat'].mean()
fatstd = df['fat'].std()

fatoutlier = df[df['fat']>fatmean+fatstd]
fatoutlier
```

```python
carbmean = df['carbs'].mean()
carbstd = df['carbs'].std()

carboutlier = df[df['carbs']>carbmean+carbstd]
carboutlier
```

```python
proteinmean = df['protein'].mean()
proteinstd = df['protein'].std()

proteinoutlier = df[df['protein']>proteinmean+proteinstd]
proteinoutlier
```

```python
calmean = df['calories'].mean()
calstd = df['calories'].std()

caloutlier = df[df['calories']>calmean+calstd]
caloutlier
```

```python
df = df[~(df['calories']>calmean+calstd)]
df
```

```python
df.boxplot(numeric)
```

```python
plt.subplots(figsize=(10,10))
sns.heatmap(df.corr());
```

```python
sns.pairplot(df)
```
